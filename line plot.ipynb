## 1. Linienplot 

### Definition:
Zeigt Daten an als eine Reihe von Datenpunkten, die durch gerade Linien verbunden sind.
### Verwendungen:
- line plots zeigen Trends, wie z.B. Schwankungen an der Börse, Temperature im Laufe der Zeit usw.
- an der x-achse werden kontinuierliche unabhängige Variablen eingesetzt wie Alter, Zeit...
- Daten werden verglichen nach diesen Variablen (x-achse)
- line plots veranschaulichen Ursache-Wirkung-Beziehungen z.B. Veränderungen im Umsatz aufgrund des Marketingsbudgets
- Der Auswahl der Skala spielt eine Rolle, je nach Absicht der Visualisierung: wenn es um relative Unterschiede geht, dann müssen wir eine kleine Skala auswählen anderfalls es ist empfehlungswert eine größere Skala anzuwenden

### Übersicht über Daten:
Die Daten umfassen den Gender Wage Gap bzw. den Unterschied im durchschnittlichen Einkommen zwischen Männer und Frauen in verschiedenen Ländern über die Jahre von 1981 bis 2016. Ein Gender Wage Gap von 12% bedeutet, dass Frauen im Durchschnitt 12% weniger  als Männer verdienen, während -15 % bedeutet, dass Frauen im Durchschnitt 15% mehr als Männer verdienen

```python
df = pd.read_csv('https://raw.githubusercontent.com/solainv/Line-Plot/main/gendergapinaverage.csv')
df.head()
```

```python
df.shape
```

```python
df['country'].value_counts()
```

```python
brazil = df[df['country']=='Brazil']
brazil = brazil[['Year', 'Gender wage gap %']]
```

```python
# Liniendiagramm für den Gender Wage Gap in Brasilien erstellen
brazil.plot(kind='line', x='Year', y='Gender wage gap %', figsize=(10, 6))
```
